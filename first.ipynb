```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
```

```python
df.head()
```

```python
plt.scatter(df['age'], df['fare'])
```

```python
df1 = df.groupby("survived")['fare'].sum()
df1
```

```python
plt.bar(["Not Survived", "Survived"], df1, color = ['red', 'blue'])
```

```python
df.head()
```

```python
df3 = df.groupby('survived')['fare'].sum()
df3
```

```python
plt.bar(['Not survived','survived'],df3,color=['red','yellow'])
```

```python
plt.scatter(df['pclass'], df['fare'], color='red')
plt.title('Pclass vs Fare')
plt.xlabel('Pclass')
plt.ylabel('Fare')
plt.show()
```
